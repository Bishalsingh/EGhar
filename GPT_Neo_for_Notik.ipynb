```python
from transformers import GPTNeoForCausalLM, GPT2Tokenizer
from transformers import TrainingArguments
from transformers import Trainer
import torch

device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(device)

mname = "EleutherAI/gpt-neo-1.3B"
tokenizer = GPT2Tokenizer.from_pretrained(mname)
model = GPTNeoForCausalLM.from_pretrained(mname).cuda()
model.to(device)

prompt = (
    "In a beautiful valley, an old man lived"
)

inputs = tokenizer(prompt, return_tensors="pt").input_ids.to(device)
gen_tokens = model.generate(
    inputs, 
    do_sample = True,
    temperature = 0.9,
    max_length = 100
)
text = tokenizer.batch_decode(gen_tokens)[0]
print(text)

del model 
torch._C._cuda_emptyCache()
```

Let's create a dataset class for this training dataset

```python
from torch.utils.data import Dataset

class OffersDataset(Dataset):
  def __init__(self, offer_name_list, first_description_list, tokenizer, max_length):
    self.input_ids = []
    self.attn_masks = []
    self.labels = []
    for offer, label in zip(offer_name_list, first_description_list):
      # prepare the text
      prep_text = f'<startoftext>Offer title : {offer}\n first description: {label}<endoftext>'
      encodings_dict = tokenizer(prep_text, truncation = True, max_length = max_length, padding = "max_length")
      label_encoded = tokenizer.encode(label)
      # append to list
      self.input_ids.append(torch.tensor(encodings_dict['input_ids']))
      self.attn_masks.append(torch.tensor(encodings_dict['attention_mask']))
      self.labels.append(torch.tensor(label_encoded))

  def __len__(self):
    return len(self.input_ids)

  def __getitem__(self, idx):
    return self.input_ids[idx], self.attn_masks[idx], self.labels[idx]


def load_offer_dataset(tokenizer):
   pass
```

```python
# Let's create a function to load the dataset

from sklearn.model_selection import train_test_split

def load_dataset(tokenizer, dataset):
  X_train, X_test, y_train, y_test = \
                train_test_split(dataset['offer_name'].tolist(), dataset['first_description'].tolist(), 
                                                       shuffle=True, test_size=0.05, random_state=1)
  train_dataset = OffersDataset(X_train, y_train, tokenizer=tokenizer, max_length = 512)
  return train_dataset, (X_test, y_test)

```

Let's fine tune the model

```python

tokenizer_notik = GPT2Tokenizer.from_pretrained(mname, bos_token='<startoftext>', eos_token = '<endoftext>', pad_token = '<pad>')
model_notik = GPTNeoForCausalLM.from_pretrained(mname).cuda()
train_dataset, test_dataset = load_dataset(tokenizer_notik, site_offers_df)

# Let's train
training_args = TrainingArguments(output_dir = 'results', num_train_epochs=2, logging_steps=10, logging_dir = 'logs')

# Start Training
Trainer(model=model_notik, args=training_args, train_dataset=train_dataset).train()
```
